```python
import pandas as pd 
 

relatorio_df = pd.read_excel (r'CONFERENCIA DOS IMPOSTOS.xlsx')

pd.set_option('display.max_columns', None) #Mostrar todas as colunas
#pd.set_option('display.max_rows', None) #Mostrar todas as linhas 

display(relatorio_df)
```

```python
#Tornando o tipo das colunas em strings 
relatorio_df['CFOP'] = relatorio_df['CFOP'].astype('str')

#Filtrando os codigos pelo primeiro indice, pelo primeiro digito.
cod5 = relatorio_df[relatorio_df['CFOP'].apply(lambda x:x[0])=='5']
cod6 = relatorio_df[relatorio_df['CFOP'].apply(lambda x:x[0])=='6']

cod1 = relatorio_df[relatorio_df['CFOP'].apply(lambda x:x[0])=='1']
cod2 = relatorio_df[relatorio_df['CFOP'].apply(lambda x:x[0])=='2']


```

```python
#Calculo ICMS - TOtal de Débitos

total_debitos5_icms = cod5['Imp. ICMS'].sum()
total_debitos6_icms = cod6['Imp. ICMS'].sum()

total_debitos_icms = total_debitos5_icms + total_debitos6_icms

print('{:,}'.format(total_debitos_icms))
```

```python
#Calculo ICMS - TOtal de Créditos

total_credito1_icms = cod1['Imp. ICMS'].sum()
total_credito2_icms = cod2['Imp. ICMS'].sum()

total_credito_icms = total_credito1_icms + total_credito2_icms

print('{:,}'.format(total_credito_icms))
```

```python
#Calculo IPI - Total de Débito

total_debito1_ipi = cod5['Imp. IPI'].sum()
total_debito2_ipi = cod6['Imp. IPI'].sum()

total_debito_ipi = total_debito1_ipi + total_debito2_ipi

print('{:,}'.format(total_debito_ipi))
```

```python
#Calculo IPI - Total de Créditos

total_credito1_ipi = cod1['Imp. IPI'].sum()
total_credito2_ipi = cod2['Imp. IPI'].sum()

total_credito_ipi = total_credito1_ipi + total_credito2_ipi

print('{:,}'.format(total_credito_ipi))
```

```python
#Calculo PIS - Faturamento Bruto 

CFOP1 = relatorio_df[relatorio_df['CFOP'] =='5105']
CFOP11 = CFOP1['Valor Contábil'].sum()

CFOP2 = relatorio_df[relatorio_df['CFOP'] =='5106']
CFOP22 = CFOP2['Valor Contábil'].sum()

CFOP3 = relatorio_df[relatorio_df['CFOP'] =='5403']
CFOP33 = CFOP3['Valor Contábil'].sum()

CFOP4 = relatorio_df[relatorio_df['CFOP'] =='6105']
CFOP44 = CFOP4['Valor Contábil'].sum()

CFOP5 = relatorio_df[relatorio_df['CFOP'] =='6106']
CFOP55 = CFOP5['Valor Contábil'].sum()

CFOP6 = relatorio_df[relatorio_df['CFOP'] =='6107']
CFOP66 = CFOP6['Valor Contábil'].sum()

CFOP7 = relatorio_df[relatorio_df['CFOP'] =='6108']
CFOP77 = CFOP7['Valor Contábil'].sum()

CFOP8 = relatorio_df[relatorio_df['CFOP'] =='6403']
CFOP88 = CFOP8['Valor Contábil'].sum()


faturamento_bruto_pis = CFOP11 + CFOP22 + CFOP33 + CFOP44 + CFOP55 + CFOP66 + CFOP77 + CFOP88

print('{:,}'.format(faturamento_bruto_pis))
```

```python
#Calculo PIS - ICMS 

CFOP1 = relatorio_df[relatorio_df['CFOP'] =='5105']
CFOP11 = CFOP1['Imp. ICMS'].sum()

CFOP2 = relatorio_df[relatorio_df['CFOP'] =='5106']
CFOP22 = CFOP2['Imp. ICMS'].sum()

CFOP3 = relatorio_df[relatorio_df['CFOP'] =='5403']
CFOP33 = CFOP3['Imp. ICMS'].sum()

CFOP4 = relatorio_df[relatorio_df['CFOP'] =='6105']
CFOP44 = CFOP4['Imp. ICMS'].sum()

CFOP5 = relatorio_df[relatorio_df['CFOP'] =='6106']
CFOP55 = CFOP5['Imp. ICMS'].sum()

CFOP6 = relatorio_df[relatorio_df['CFOP'] =='6107']
CFOP66 = CFOP6['Imp. ICMS'].sum()

CFOP7 = relatorio_df[relatorio_df['CFOP'] =='6108']
CFOP77 = CFOP7['Imp. ICMS'].sum()

CFOP8 = relatorio_df[relatorio_df['CFOP'] =='6403']
CFOP88 = CFOP8['Imp. ICMS'].sum()

ICMS_pis = CFOP11 + CFOP22 + CFOP33 + CFOP44 + CFOP55 + CFOP66 + CFOP77 + CFOP88

print('{:,}'.format(ICMS_pis))
```

```python
#Calculo PIS - ICMS ST

CFOP1 = relatorio_df[relatorio_df['CFOP'] =='5105']
CFOP11 = CFOP1['ICMS ST'].sum()

CFOP2 = relatorio_df[relatorio_df['CFOP'] =='5403']
CFOP22 = CFOP2['ICMS ST'].sum()

CFOP3 = relatorio_df[relatorio_df['CFOP'] =='6105']
CFOP33 = CFOP3['ICMS ST'].sum()

CFOP4 = relatorio_df[relatorio_df['CFOP'] =='6107']
CFOP44 = CFOP4['ICMS ST'].sum()

CFOP5 = relatorio_df[relatorio_df['CFOP'] =='6108']
CFOP55 = CFOP5['ICMS ST'].sum()

CFOP6 = relatorio_df[relatorio_df['CFOP'] =='6403']
CFOP66 = CFOP6['ICMS ST'].sum()


ICMS_ST_pis = CFOP11 + CFOP22 + CFOP33 + CFOP44 + CFOP55 + CFOP66

print('{:,}'.format(ICMS_ST_pis))
```

```python
#Calculo PIS - IPI

CFOP1 = relatorio_df[relatorio_df['CFOP'] =='5105']
CFOP11 = CFOP1['Imp. IPI'].sum()

CFOP2 = relatorio_df[relatorio_df['CFOP'] =='5403']
CFOP22 = CFOP2['Imp. IPI'].sum()

CFOP3 = relatorio_df[relatorio_df['CFOP'] =='6105']
CFOP33 = CFOP3['Imp. IPI'].sum()

CFOP4 = relatorio_df[relatorio_df['CFOP'] =='6107']
CFOP44 = CFOP4['Imp. IPI'].sum()

CFOP5 = relatorio_df[relatorio_df['CFOP'] =='6108']
CFOP55 = CFOP5['Imp. IPI'].sum()

CFOP6 = relatorio_df[relatorio_df['CFOP'] =='6403']
CFOP66 = CFOP6['Imp. IPI'].sum()


IPI_pis = CFOP11 + CFOP22 + CFOP33 + CFOP44 + CFOP55 + CFOP66

print('{:,}'.format(IPI_pis))
```

```python
#Calculo de PIS - Base de Calculo PIS 

Base_de_calculo_pis = faturamento_bruto_pis - ICMS_pis - ICMS_ST_pis - IPI_pis

print('{:,}'.format(Base_de_calculo_pis))
```

```python
#Calculo de PIS - Crédito PIS 


CST_PIS1 = cod1[cod1['CST PIS']==50]
credito_pis1 = CST_PIS1['Imposto PIS'].sum()

CST_PIS2 = cod2[cod2['CST PIS']==50]
credito_pis2 = CST_PIS2['Imposto PIS'].sum()

credito_pis = credito_pis1 + credito_pis2

print('{:,}'.format(credito_pis))
```

```python
#Calculo de PIS - Saldo a pagar 

saldo_a_pagar_pis = (Base_de_calculo_pis * 1.65 / 100) - credito_pis
print('{:,}'.format(saldo_a_pagar_pis))
```

```python
# Calculo do COFINS 

faturamento_bruto_cofins = faturamento_bruto_pis
ICMS_cofins = ICMS_pis
ICMS_ST_cofins = ICMS_ST_pis
IPI_cofins = IPI_pis
Base_de_calculo_cofins = Base_de_calculo_pis

#Calculo do COFINS - Crédito COFINS 

CST_COFINS1 = cod1[cod1['CST PIS']==50]
credito_cofins1 = CST_COFINS1['Imp. COFINS'].sum()

CST_COFINS2 = cod2[cod2['CST PIS']==50]
credito_cofins2 = CST_COFINS2['Imp. COFINS'].sum()

credito_cofins = credito_cofins1 + credito_cofins2

print('{:,}'.format(credito_cofins))
```

```python
# Calculo do COFINS - Saldo a pagar 

saldo_a_pagar_cofins = (Base_de_calculo_cofins * 7.60 / 100) - credito_cofins

print('{:,}'.format(saldo_a_pagar_cofins))
```

```python
print('- Calculo ICMS -\n') 
print('Total de Débitos: {:,.2f}'.format(total_debitos_icms))
print('Total de Crébito: {:,.2f}'.format(total_credito_icms))
print('Saldo Mês Anterior:')
print('Saldo: \n\n')

print('- Calculo IPI -\n') 
print('Total de Débitos: {:,.2f}'.format(total_debito_ipi))
print('Total de Crébito: {:,.2f}'.format(total_credito_ipi))
print('Saldo Mês Anterior:')
print('Saldo: \n\n')

print('- Calculo PIS -\n') 
print('Faturamento Bruto: {:,.2f}'.format(faturamento_bruto_pis))
print('ICMS: {:,.2f}'.format(ICMS_pis))
print('ICMS ST: {:,.2f}'.format(ICMS_ST_pis))
print('IPI: {:,.2f}\n'.format(IPI_pis))
print('Base de Calculo PIS: {:,.2f}\n'.format(Base_de_calculo_pis))
print('Crédito PIS: {:,.2f}'.format(credito_pis))
print('Saldo a Pagar: {:,.2f}\n\n'.format(saldo_a_pagar_pis))

print('- Calculo COFINS -\n') 
print('Faturamento Bruto: {:,.2f}'.format(faturamento_bruto_cofins))
print('ICMS: {:,.2f}'.format(ICMS_cofins))
print('ICMS ST: {:,.2f}'.format(ICMS_ST_cofins))
print('IPI: {:,.2f}\n'.format(IPI_cofins))
print('Base de Calculo COFINS: {:,.2f}\n'.format(Base_de_calculo_cofins))
print('Crédito PIS: {:,.2f}'.format(credito_cofins))
print('Saldo a Pagar: {:,.2f}'.format(saldo_a_pagar_cofins))
```

```python
arquivo_df = pd.read_excel(r'Relatorio.xlsx')

arquivo_df.at[3, 'C'] = total_debitos_icms
arquivo_df.at[4, 'C'] = total_credito_icms

arquivo_df.at[11, 'C'] = total_debito_ipi
arquivo_df.at[12, 'C'] = total_credito_ipi

arquivo_df.at[18, 'C'] = faturamento_bruto_pis
arquivo_df.at[19, 'C'] = ICMS_pis
arquivo_df.at[20, 'C'] = ICMS_ST_pis
arquivo_df.at[21, 'C'] = IPI_pis

arquivo_df.at[23, 'C'] = Base_de_calculo_pis

arquivo_df.at[25, 'C'] = credito_pis
arquivo_df.at[26, 'C'] = saldo_a_pagar_pis

arquivo_df.at[30, 'C'] = faturamento_bruto_cofins
arquivo_df.at[31, 'C'] = ICMS_cofins
arquivo_df.at[32, 'C'] = ICMS_ST_cofins
arquivo_df.at[33, 'C'] = IPI_cofins

arquivo_df.at[35, 'C'] = Base_de_calculo_cofins

arquivo_df.at[37, 'C'] = credito_cofins
arquivo_df.at[38, 'C'] = saldo_a_pagar_cofins


arquivo_df.to_excel(r'Relatorio Atualizado.xlsx', sheet_name='Relatorio_final', na_rep=' ', index=False)
```
